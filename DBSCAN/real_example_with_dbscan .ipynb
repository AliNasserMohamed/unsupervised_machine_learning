```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
X, _ = make_moons(n_samples=200, noise=0.05)
```

```python
k = 10
neigh = NearestNeighbors(n_neighbors=k)
nbrs = neigh.fit(X)
distances, _ = nbrs.kneighbors(X)
distances = distances[:, -1]

# Sort and plot the distances
distances = sorted(distances)
plt.plot(distances)
plt.xlabel("Samples")
plt.ylabel("Distance to {}th nearest neighbor".format(k))
plt.title("Elbow Method for Optimal Epsilon")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.17, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs, make_circles
from sklearn.cluster import DBSCAN

# Create a function to visualize DBSCAN clustering on different datasets
def visualize_dbscan(X, eps, min_samples, title, position):
    plt.subplot(2, 3, position)  # Adjust the grid layout to 2 rows and 3 columns
    
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    clusters = dbscan.fit_predict(X)
    print(clusters)
    
    plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
    plt.title(title)

# Create synthetic datasets
X_moons, _ = make_moons(n_samples=200, noise=0.05)
X_blobs, _ = make_blobs(n_samples=200, centers=3, cluster_std=1.0)
X_circles, _ = make_circles(n_samples=200, factor=0.5, noise=0.05)
X_custom, _ = make_blobs(n_samples=200, centers=4, cluster_std=0.7)  # Add another dataset

# Create two more synthetic datasets
X_squares, _ = make_blobs(n_samples=200, centers=5, cluster_std=0.6)  # Add another dataset
X_triangles, _ = make_blobs(n_samples=200, centers=2, cluster_std=0.3)  # Add another dataset

# Set the subplot layout
plt.figure(figsize=(18, 12))  # Adjust the figure size

# Visualize DBSCAN on different datasets
eps = 0.4
min_samples = 5

visualize_dbscan(X_moons, eps, min_samples, 'DBSCAN on Moons', 1)
visualize_dbscan(X_blobs, eps, min_samples, 'DBSCAN on Blobs', 2)
visualize_dbscan(X_circles, eps, min_samples, 'DBSCAN on Circles', 3)
visualize_dbscan(X_custom, eps, min_samples, 'DBSCAN on Custom Data', 4)
visualize_dbscan(X_squares, eps, min_samples, 'DBSCAN on Squares', 5)  # Add another dataset at position 5
visualize_dbscan(X_triangles, eps, min_samples, 'DBSCAN on Triangles', 6)  # Add another dataset at position 6

# Show the plots
plt.show()
```
